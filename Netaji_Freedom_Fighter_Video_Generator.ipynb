```python
import requests

url = "https://en.wikipedia.org/w/api.php"

params = {
    "action": "query",
    "format": "json",
    "titles": "Subhas Chandra Bose",
    "prop": "pageimages",
    "pithumbsize": 500
}

# Add headers (important!)
headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, params=params, headers=headers)

print("Status Code:", response.status_code)

if response.status_code == 200:
    data = response.json()

    pages = data["query"]["pages"]
    page = list(pages.values())[0]

    image_url = page["thumbnail"]["source"]

    print("✅ API Connected!")
    print("Image URL:", image_url)

    # Download image
    netaji_img = requests.get(image_url, headers=headers).content

    with open("netaji_img", "wb") as f:
        f.write(netaji_img)

    print("✅ Netaji image downloaded!")

else:
    print("❌ API failed")
```

```python
W, H         = 1280, 720
FPS          = 25
DURATION     = 10
TOTAL_FRAMES = FPS * DURATION

SAFFRON = (255, 153,  51)
WHITE   = (255, 255, 255)
GOLD    = (255, 210,   0)
GREEN   = ( 19, 136,   8)
NAVY    = ( 10,  15,  55)
MAROON  = ( 28,   8,   8)
CREAM   = (255, 240, 200)

print(f"✅ Total frames: {TOTAL_FRAMES}")
```

```python
def get_font(size):
    paths = [
        'C:/Windows/Fonts/arialbd.ttf',
        'C:/Windows/Fonts/arial.ttf',
        '/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf',
        '/Library/Fonts/Arial Bold.ttf',
    ]
    for p in paths:
        if os.path.exists(p):
            return ImageFont.truetype(p, size)
    return ImageFont.load_default()

def make_gradient():
    arr = np.zeros((H, W, 3), dtype=np.uint8)
    for y in range(H):
        t = y / H
        arr[y] = [int(NAVY[i]*(1-t) + MAROON[i]*t) for i in range(3)]
    return Image.fromarray(arr)

def draw_tricolour(img, y, h=9):
    d = ImageDraw.Draw(img)
    s = h // 3
    d.rectangle([0, y,     W, y+s],   fill=SAFFRON)
    d.rectangle([0, y+s,   W, y+2*s], fill=WHITE)
    d.rectangle([0, y+2*s, W, y+h],   fill=GREEN)

def draw_shadow(draw, pos, text, font, color):
    draw.text((pos[0]+2, pos[1]+2), text, font=font, fill=(0,0,0))
    draw.text(pos, text, font=font, fill=color)

def tsize(draw, text, font):
    b = draw.textbbox((0,0), text, font=font)
    return b[2]-b[0], b[3]-b[1]

print("✅ Helpers ready!")
```

```python
IMAGE_PATH = 'netaji.jpg'  # ⚠️ make sure netaji.jpg is in same folder

netaji_img = Image.open(IMAGE_PATH).convert('RGB')
print(f"✅ Image loaded! Size: {netaji_img.size}")
display(netaji_img)
```

```python
TARGET_H = 430
TARGET_W = int(netaji_img.width * TARGET_H / netaji_img.height)
portrait = netaji_img.resize((TARGET_W, TARGET_H), Image.LANCZOS)
portrait = ImageEnhance.Contrast(portrait).enhance(1.2)
portrait = ImageEnhance.Sharpness(portrait).enhance(1.4)
portrait = ImageEnhance.Brightness(portrait).enhance(1.05)

def add_border(img, size, color):
    w, h = img.size
    c = Image.new('RGB', (w+2*size, h+2*size), color)
    c.paste(img, (size, size))
    return c

portrait = add_border(portrait, 5, SAFFRON)
portrait = add_border(portrait, 3, WHITE)
portrait = add_border(portrait, 6, GOLD)

PW, PH = portrait.size
PX = W - PW - 55
PY = (H - PH) // 2 - 20

print(f"✅ Portrait ready: {PW}x{PH}")
display(portrait)
```

```python
F_BIG   = get_font(62)
F_TITLE = get_font(50)
F_QUOTE = get_font(28)
F_BODY  = get_font(32)
F_SMALL = get_font(22)

BG    = make_gradient()
DIV_X = PX - 75

print("✅ Fonts and background ready!")
```

```python
SLIDES = [ (0.0, 2.5, 'NETAJI', 'Subhas Chandra Bose',
     ['23 Jan 1897 – 18 Aug 1945', 'Hero of Indian Independence']),
    (2.5, 5.0, 'Fearless Leader',
     '"Give me blood, I will give you freedom!"',
     ['Greatest orator & visionary', 'Rallied millions to the cause']),
    (5.0, 7.5, 'Azad Hind Fauj', 'Indian National Army (INA)',
     ['Founded INA to liberate India',
      'Marched under banner of Jai Hind']),
    (7.5, 10.0, 'Eternal Inspiration', 'A legend who lives forever',
     ['Symbol of sacrifice & courage', 'JAI HIND !']),
]

def get_slide(fi):
    t = fi / FPS
    for s in SLIDES:
        if s[0] <= t < s[1]:
            return s
    return SLIDES[-1]

print(f"✅ {len(SLIDES)} slides ready!")
```

```python
def render_frame(fi):
    s0, s1, headline, subhead, bullets = get_slide(fi)
    t       = fi / FPS
    dur     = s1 - s0
    local_t = t - s0
    fade    = min(0.4, dur * 0.18)

    if local_t < fade:
        alpha = local_t / fade
    elif local_t > dur - fade:
        alpha = (dur - local_t) / fade
    else:
        alpha = 1.0
    alpha = max(0.0, min(1.0, alpha))

    frame = BG.copy()
    d     = ImageDraw.Draw(frame)

    draw_tricolour(frame, y=0,   h=9)
    draw_tricolour(frame, y=H-9, h=9)
    d.line([(DIV_X,25),(DIV_X,H-25)], fill=GOLD, width=2)
    d.line([(30,115),(DIV_X-15,115)], fill=GOLD, width=2)

    frame.paste(portrait, (PX, PY))

    cap = 'Netaji Subhas Chandra Bose'
    cw, _ = tsize(d, cap, F_SMALL)
    draw_shadow(d, (PX+(PW-cw)//2, PY+PH+8), cap, F_SMALL, SAFFRON)

    PAD = 38
    TW  = DIV_X - PAD - 15
    hfont = F_BIG if s0 == 0.0 else F_TITLE
    hy = 130
    for line in textwrap.wrap(headline, 18):
        lw, lh = tsize(d, line, hfont)
        draw_shadow(d, (PAD+max(0,(TW-lw)//2), hy), line, hfont, GOLD)
        hy += lh + 8

    hy += 10
    for line in textwrap.wrap(subhead, 30):
        lw, lh = tsize(d, line, F_QUOTE)
        draw_shadow(d, (PAD+max(0,(TW-lw)//2), hy), line, F_QUOTE, CREAM)
        hy += lh + 5

    hy += 14
    d.line([(PAD, hy),(DIV_X-15, hy)], fill=SAFFRON, width=2)
    hy += 20

    for b in bullets:
        d.ellipse([(PAD, hy+12),(PAD+9, hy+21)], fill=SAFFRON)
        for line in textwrap.wrap(b, 28):
            _, lh = tsize(d, line, F_BODY)
            draw_shadow(d, (PAD+18, hy), line, F_BODY, WHITE)
            hy += lh + 5
        hy += 12

    si = next(i for i,s in enumerate(SLIDES) if s[0]==s0)
    for i in range(len(SLIDES)):
        cx = W//2 - 40 + i*20
        d.ellipse([(cx-5,H-27),(cx+5,H-17)],
                  fill=SAFFRON if i==si else (70,70,70))

    d.rectangle([0,H-3,int(W*t/DURATION),H], fill=SAFFRON)

    if alpha < 1.0:
        avg = tuple((NAVY[i]+MAROON[i])//2 for i in range(3))
        overlay = Image.new('RGB',(W,H), avg)
        frame   = Image.blend(overlay, frame, alpha)

    return np.array(frame)

print("✅ Renderer ready!")
```

```python
print(f"⏳ Rendering {TOTAL_FRAMES} frames...")

frames = []
for i in range(TOTAL_FRAMES):
    frames.append(render_frame(i))
    if (i+1) % 50 == 0:
        print(f"   {i+1}/{TOTAL_FRAMES} done...")

# Export using imageio (no moviepy needed!)
writer = imageio.get_writer(
    'netaji_introduction.mp4',
    fps=FPS,
    codec='libx264',
    quality=8
)
for f in frames:
    writer.append_data(f)
writer.close()

print("🎉 Video saved → netaji_introduction.mp4")
```
